```python
import pandas as pd
import os

# Ruta hacia los archivos crudos
path = '../data/raw/'

# Lista de los años que vamos a leer
years = ['2015', '2016', '2017', '2018', '2019']

# Aquí guarda nuestros 5 dataframes 
dataframes = {}

# Ciclo para leer cada archivo y mostrar sus columnas
for year in years:
    file_name = f"{path}{year}.csv"
    
    # Leemos el CSV
    df = pd.read_csv(file_name)
    dataframes[year] = df
    
    # Imprime el nombre del año y la lista de sus columnas
    print(f" Columnas del archivo {year}.csv:")
    print(df.columns.tolist())
    print("-" * 50)
```

```python
# 1. Definimos los diccionarios para traducir las columnas de cada año a un estándar
cols_2015_2016 = {
    'Country': 'country', 'Happiness Rank': 'rank', 'Happiness Score': 'score',
    'Economy (GDP per Capita)': 'gdp', 'Family': 'social_support', 
    'Health (Life Expectancy)': 'health', 'Freedom': 'freedom', 
    'Trust (Government Corruption)': 'corruption', 'Generosity': 'generosity'
}

cols_2017 = {
    'Country': 'country', 'Happiness.Rank': 'rank', 'Happiness.Score': 'score',
    'Economy..GDP.per.Capita.': 'gdp', 'Family': 'social_support', 
    'Health..Life.Expectancy.': 'health', 'Freedom': 'freedom', 
    'Trust..Government.Corruption.': 'corruption', 'Generosity': 'generosity'
}

cols_2018_2019 = {
    'Country or region': 'country', 'Overall rank': 'rank', 'Score': 'score',
    'GDP per capita': 'gdp', 'Social support': 'social_support', 
    'Healthy life expectancy': 'health', 'Freedom to make life choices': 'freedom', 
    'Perceptions of corruption': 'corruption', 'Generosity': 'generosity'
}

# 2. procesa cada dataframe que cargamos antes
dataframes_limpios = []

for year, df in dataframes.items():
    # Asignamos el diccionario correcto según el año
    if year in ['2015', '2016']:
        df_rename = df.rename(columns=cols_2015_2016)
    elif year == '2017':
        df_rename = df.rename(columns=cols_2017)
    else:
        df_rename = df.rename(columns=cols_2018_2019)
        
    # Agrega una columna para saber de qué año viene el dato
    df_rename['year'] = year
    
    # Nos quedamos solamente con las columnas estandarizadas, filtramos la basura
    columnas_finales = ['country', 'year', 'rank', 'score', 'gdp', 'social_support', 'health', 'freedom', 'generosity', 'corruption']
    df_limpio = df_rename[columnas_finales]
    
    # Lo guardamos en nuestra lista de dataframes limpios
    dataframes_limpios.append(df_limpio)

# 3. unimos los 5 archivos uno debajo del otro
df_final = pd.concat(dataframes_limpios, ignore_index=True)

# 4. Exportamos este grande archivo limpio a nuestra carpeta de procesados
df_final.to_csv('../data/processed/clean_happiness_data.csv', index=False)

print("Limpieza terminada con éxito")
print(f"Nuestro dataset final tiene {df_final.shape[0]} filas y {df_final.shape[1]} columnas.")
df_final.head() # Muestra las primeras 5 filas para que veamos cómo quedó
```
